```python
# Questão 1

print('Tamanho de camiseta\n')

print('Olá! Vamos tirar suas medidas?\n')

try:
  larg = float(input('Sua largura:\n'))
  alt = float(input('Sua altura:\n'))

  if (larg >= 20 and larg <= 25 and alt >= 25 and alt <= 30):
      print('Tamanho P.')
  elif (larg >= 26 and larg <= 30 and alt >= 31 and alt <= 35):
      print('Tamanho M.')
  elif (larg >= 31 and larg <= 35 and alt >= 36 and alt <= 40):
      print('Tamanho G.')
  elif (larg >= 36 and larg <= 40 and alt >= 41 and alt <= 45):
      print('Tamanho XG.')
  elif (larg >= 41 and larg <= 45 and alt >= 46 and alt <= 50):
      print('Tamanho XXG.')
  elif (larg >= 46 and larg <= 50 and alt >= 51 and alt <= 55):
      print('Tamanho 3X.')
  elif (larg >= 51 and larg <= 55 and alt >= 56 and alt <= 60):
      print('Tamanho 4X')
  else:
    print('Tamanho inexistente.')
except:
  print('Digite apenas números.')

```

```python
# Questão 2

print('Quantidade de calorias ingeridas por dia\n')

i = 0
maior = -9999
menor = 9999
soma = 0
pes = 10

while(i < pes):
  try:
    cal = int(input('Quantidade de calorias:\n'))
    soma = soma + cal
    if (cal > maior):
      maior = cal
    if (cal < menor):
      menor = cal
    i = i + 1
  except:
    print('Digite apenas números.')

print(f'\nA média de calorias ingeridas é igual a: {round(soma/pes,2)}')
print(f'Quantidade máxima de calorias: {maior}\nQuantidade mínima de calorias: {menor}')
```

```python
# Questão 3

print('Quantidade de calorias ingeridas por dia\n')

i = 0
maior = -9999
menor = 9999
soma = 0
pes = int(input('De quantas pessoas pretende calcular?\n'))
cont = 0

while(i < pes):
  try:
    cal = int(input(f'Quantidade de calorias da pessoa [{i+1}]:\n'))
    soma = soma + cal
    if (cal > maior):
      maior = cal
    if (cal < menor):
      menor = cal
    if (cal > 2000):
      cont = cont + 1
    i = i + 1
  except:
    print('Digite apenas números.')

print(f'\nA média de calorias ingeridas é igual a: {round(soma/pes,2)}')
print(f'\nConsomem mais de 2000 calorias: {cont}')
print(f'Quantidade máxima de calorias: {maior}\nQuantidade mínima de calorias: {menor}')
```
